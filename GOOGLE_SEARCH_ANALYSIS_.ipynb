```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
trends = TrendReq()
```

# ***CREATE DATAFRAME FOR TOP 10 CONTRIES WHICH SEARCH  MACHINE LEARNING ON GOOGLE ***

---



```python
trends.build_payload(kw_list=["Machine Learning"])
data = trends.interest_by_region()
data = data.sort_values(by="Machine Learning", ascending=False)
data = data.head(10)
print(data)
```

## ACCORDING TO QUERIES CHINA HAS MORE SEARCHES IN MACHINE LEARNING ;
# ***DATA VISUALIZATION ***

---



```python
data.reset_index().plot(x="geoName", 
                        y="Machine Learning", 
                        figsize=(15,12), kind="bar",color='orange')
plt.style.use('fivethirtyeight')
plt.show()
```

# **HOW TREND SEARCHES OF MACHINE LEARNING VARIES WITH YEAR ?**

---



```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Machine Learning'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(15, 12))
data['Machine Learning'].plot(color='green')
plt.style.use('fivethirtyeight')
plt.title('Total Google Searches for Machine Learning', 
          fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```

# **DATA VISUALIZATION**

---



Create PIE CHART 

```python
trends = TrendReq(hl='en-US', tz=360)
trends.build_payload(kw_list=["Billionaries"])
data = trends.interest_by_region()
data = data.sort_values(by="Billionaries", ascending=False)
data = data.head(10)

fig, ax = plt.subplots(figsize=(22,20))
ax.pie(data["Billionaries"], labels=data.index, autopct='%1.1f%%',
       shadow=True, startangle=120)
ax.set_title('Regions with highest interest in Billionaries', fontweight='bold')
plt.show()
```

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Billionaries'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(15, 12))
data['Billionaries'].plot(color='black')
plt.style.use('fivethirtyeight')
plt.title('Total Google Searches for Billionaries', 
          fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```
